# SVD
- singular value decomposition

## $UDV^T$
- U : m x m matrix
- D : m x n diagonal matrix
- V : n x n matrix

# Seperable

```python
import numpy as np
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage
 
x, y = np.meshgrid(np.linspace(-1, 1, 50), np.linspace(-1, 1, 50))
gauss = np.exp(-5 * (x * x + y * y))
plt.imshow(gauss)
plt.colorbar()
```

```python
box = np.array(np.logical_and(np.abs(x) < 0.7, np.abs(y) < 0.7),dtype='float64') 
plt.imshow(box)
plt.colorbar()
```

```python
U, D, V = np.linalg.svd(box)
print(f"box : {box.shape}")
print(f"D : {D.shape}")
print(f"U : {U.shape}")
print(f"V : {V.shape}")
```

```python
new_box = D[0] * np.outer(U[:, 0], V[0, :])
plt.imshow(new_box)
```

# Non Seperable

```python
circle = np.array(x*x+y*y < 0.8, dtype='float64')

plt.imshow(circle, cmap='plasma')
plt.colorbar()
```

```python
U, D, V = np.linalg.svd(circle)
print(f"circle : {box.shape}")
print(f"D : {D.shape}")
print(f"U : {U.shape}")
print(f"V : {V.shape}")
```

```python
new_circle = D[0] * np.outer(U[:, 0], V[0, :])
plt.imshow(new_circle, cmap='plasma')
```

```python
U, D, V = np.linalg.svd(circle)
new_circle = np.zeros_like(circle)

for i in range(0, 10):
    new_circle += D[i] * np.outer(U[:, i], V[i, :])

plt.imshow(new_circle, cmap='plasma')
plt.show()
```

# Denoising

```python
noise_circle = np.array(x*x+y*y < 0.8, dtype='float64') + np.random.random((50, 50)) * 0.3

plt.imshow(noise_circle, cmap='plasma')
plt.show()
```

```python
U, D, V = np.linalg.svd(noise_circle)
new_noise_circle = np.zeros_like(noise_circle)

for i in range(0, 9):
    new_noise_circle += D[i] * np.outer(U[:, i], V[i, :])

plt.imshow(new_noise_circle, cmap='plasma')
plt.show()
```
